# **Sorting Algorithms**

```python
import random
import time
import matplotlib.pyplot as plt
import platform
import psutil
```

```python
# Implementation of insertion sort
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

# Implementation of selection sort
def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

# Implementation of bubble sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n-1):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

# Function to generate an array of random integers
def generate_array(size):
    return [random.randint(1, 1000) for _ in range(size)]


```

```python
# Hardware specifications
cpu_info = platform.processor()
ram_info = psutil.virtual_memory().total / (1024 ** 3)  # RAM in GB

(cpu_info, ram_info)
```

```python
# Benchmarking the algorithms

# Array sizes for benchmarking
sizes = [5, 10, 20, 50, 100, 200, 500, 1000, 2000, 5000]


# Time taken for each algorithm at each size
times_insertion = []
times_selection = []
times_bubble = []

# Benchmark function
def benchmark(algorithm, arr):
    start_time = time.time()
    algorithm(arr.copy())  # Use a copy to avoid sorting an already sorted array
    return time.time() - start_time

```

```python
# Run the benchmarks
for size in sizes:
    arr = generate_array(size)

    time_insertion = benchmark(insertion_sort, arr)
    times_insertion.append(time_insertion)

    time_selection = benchmark(selection_sort, arr)
    times_selection.append(time_selection)

    time_bubble = benchmark(bubble_sort, arr)
    times_bubble.append(time_bubble)

```

```python
# Plotting the results
plt.figure(figsize=(12, 6))
plt.plot(sizes, times_insertion, label='Insertion Sort', marker='o')
plt.plot(sizes, times_selection, label='Selection Sort', marker='o')
plt.plot(sizes, times_bubble, label='Bubble Sort', marker='o')
plt.xlabel('Array Size')
plt.ylabel('Time (seconds)')
plt.title('Sorting Algorithm Performance')
plt.legend()
plt.grid(True)
plt.show()
```
